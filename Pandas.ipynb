```python
import pandas as pd
```

```python
df = pd.read_csv("Fire_Data.15240.1598195125.4438.csv")
```

```python
for i in range(len(list(df['ACQ_DATE']))):
    df['ACQ_NEWDATE'][i] = df['ACQ_DATE'][i].split(' ')[0]
```

```python
for i in range(len(list(df['ACQ_TIME']))):
    df['NEWTIME'][i] = list("%04d"%(df['ACQ_TIME'][i]))[0]+list("%04d"%(df['ACQ_TIME'][i]))[1]+':'+list("%04d"%(df['ACQ_TIME'][i]))[2]+list("%04d"%(df['ACQ_TIME'][i]))[3]
```

```python
df['DATETIME'] = pd.to_datetime(df['ACQ_NEWDATE'] + ' ' + df['NEWTIME'])
```

```python
df = df.drop(['ACQ_NEWDATE', 'NEWTIME'], axis=1)
```

```python
dffiveday = df[(df['DATETIME'] > '2019-12-01 00:00') & (df['DATETIME'] < '2019-12-05 23:59')]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
axes = fig.add_subplot(111,projection = '3d')
X = dffiveday['LONGITUDE']
Y = dffiveday['LATITUDE']
Z = dffiveday['DATETIME']
axes.scatter(X,Y,Z)
axes.set_xlabel('Longtitude')
axes.set_ylabel('Latitude')
axes.set_zlabel('Days')
plt.show()
```
